### Series 객체 

```python
import pandas as pd
import numpy as np

m2 = pd.read_csv('employee.scv')
m2
```

```python
# 컬럼접근
sr = m2['ename']   # Series 객체
sr
```

```python
# srt.contains() 메서드를 사용하여 문자열의 조건 검색
sr = m2['ename'].str.contains('k')
print(sr)      # Series
print(type(sr))
m2[sr]
```

```python
# 'deptname' 이 '영업직'인 행만 추출
sr = m2['deptname'].str.contains('영업직')

m2[sr]
```

```python
# 'addr'가 광주시인 행만 추출
sr = m2['addr'].str.contains('광주시')
m2[sr]
```

```python
# 조건식 2개 사용
# deptname 이 '영업직'이고 'addr' 가 '광주시'인 행을 추출
sr1 = m2['deptname'].str.contains('영업직')
sr2 = m2['addr'].str.contains('광주시')

m2[sr1&sr2]
```

### 누락된 데이터(결측치,NaN,NaT) 다루기

```python
df = pd.read_csv('WHO_first9cols.csv')
df.count() # 결측치를 제외한 데이터의 갯수

df2 = df[['Country','Net primary school enrolment ratio male (%)']]
df2

# pd.isnull().pd.notnull()
pd.isnull(df2)    # 결측치는 True로 반환, 나머지는 False로 반환
pd.notnull(df2)   # 결측치는 False로 반환, 나머지는 True로 반환

pd.isnull(df2).sum()   # 23: 2번 컬럼의 결측지의 갯수
pd.isnull(df).sum()    # 전체 컬럼의 결측치의 갯수


# fillna() : 결측치를 특정 값으로 채움
# 결측지를 0으로 채움 : 평균, 표준 편차에 영향
df3 = df2.fillna(0)
pd.isnull(df3).sum()   # 0 , 결측치의 갯수는 0

# replace(a,b)
df4 = df2.replace(np.NaN,0)   # fillna(0)와 동이란 결과

sr = df2[df2.columns[-1]].replace(np.nan,0)  # 한개의 컬럼만 결측지를 제거

# 결측치를 평균으로 채움 : 평균에 영향을 주지않음
df5 = df2.fillna(df2.mean())
df5

# df2.mean    # 85.698324
print(df2.describe())      # 결측치 제거 전
print(df4.describe())      # 평균값이 다름
print(df5.describe())      # 평균값이 동일
```

```python
# Series 객체의 index 사용 및 결측치 제어
index_table = ['tv','vtr','phone']
sr = pd.Series([10,20,30],index=index_table)
print(sr)
print(sr.index)

r = sr.index.str.contains('v') # index에서 문자열을 검출
sr[r]
```

```python
# Series 객체의 요소의 추가
sr2 = sr.append(pd.Series([40,50,60],index=['audio','iron','notebook']))
sr2

# 테스트를 위한 결측치를 추가
sr3 = sr2.append(pd.Series([np.nan],index=['aircon']))
print(sr3)
print(sr3.mean())  # 35, NaN 을 제외하고 평균을 구함

# a = np.array([np.nan,1,2,3])  # ndarray는 nan과 연산하면 결과값이 nan이다
# a.mean()    # nan

# 결측치를 처리(제거)
sr4 = sr3.fillna(0)
print(sr4)
print(sr4.mean())  # 30, 0으로 채우면 평균값이 감소

sr5 = sr3.fillna(sr3.mean())
print(sr5)
print(sr5.mean())  # 35, 평균으로 채우면 평균값 변동이 없음
```

### groupby()

```python
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint
seed(42)

df = pd.DataFrame({ 'Weather' : ['cold', 'hot', 'cold', 'hot', 'cold', 'hot', 'cold'],
                    'Food' : ['soup', 'soup', 'icecream', 'chocolate', 'icecream', 'icecream', 'soup'],
                    'Price' : 10 * rand(7), 
                    'Number' : randint(1, 9, 7)}) # 1~9사이 7개의 난수, 정수
df
```

```python
weather_group = df.groupby('Weather')   # DataFrameGroupBy 객체
print(weather_group)

i = 0
for name,group in weather_group:
    i += 1
    print(i,name,'\n',group)
    print('-'*30)
    
print(weather_group.first())    # 각 그룹 내 첫번째 행
print('-'*30)
print(weather_group.last())     # 각 그룹 내 마지막 행
print('-'*30)
print(weather_group.mean())
print('-'*30)
print(weather_group.count())    # 각 그룹의 총 갯수
print('-'*30)
print(weather_group.describe())  # 각 그룹의 통계 요약
```

```python
wf_group = df.groupby(['Weather','Food'])
print(wf_group)
print(wf_group.groups)   # dict

wf_group.agg([np.mean,np.sum])
```

```python
gr_df = wf_group.agg([np.mean,np.sum,np.min,np.max])
print(type(gr_df)) # DataFrame (인덱스와 컬럼이 MultiIndex인 데이터 프레임)
print(gr_df.index) # MultiIndex

print(gr_df.columns) # MultiIndex

# MultiIndex 형식의 컬럼 접근 방법 1
df1 = pd.DataFrame(gr_df['Price']['mean'])
df2 = pd.DataFrame(gr_df['Number']['mean'])
df3 = pd.concat([df1,df2],axis=1)
df3

# MultiIndex 형식의 컬럼 접근 방법 2
gr_df.loc[:,[('Price','mean'),('Number','mean')]]
```
