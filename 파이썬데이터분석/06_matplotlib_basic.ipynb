```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### 선그래프

```python
a = np.array([10,14,19,20,25])
print(a)
plt.plot(a)

plt.show()   
```

```python
# 1차 함수(방정식)의 그래프 : f(x) = a*x + b, 직선의 방정식
x = np.arange(-4.5,5.0,0.5)
y = 2*x + 0
print(x,y)

plt.plot(x,y)
plt.show
```

```python
# 2차 함수(방정식)의 그래프 : f(x) = a*x^2 + b , 포물선의 방정식
x = np.arange(-4.5,5.0,0.5)
y = 2*x**2 +0
print(x,y)

plt.plot(x,y)           # 포물선
plt.plot(x,4*x)         # 직선
plt.plot(x,-3*x)        # 직선
plt.show()
```

### 로그플롯(log plot)

```python
# 상용로그
print(np.log10(1000))  # 3.0

# 자연로그 : e : 2.71828...
print(np.log(100))
print(np.exp(4.605170185988092))  # 100.0
```

```python
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)
df   # 연도별 CPU 판매량
```

```python
years = df.index.values
counts = df['trans_count'].values

plt.plot(years,np.log(counts))   # 자연로그 사용 로그변환
plt.show()
```

```python
print(np.log1p(0))   # 1을 더한 후 log변환, log(0)이 -inf가 되지 않도록

print(np.log(1))  #
```

```python
# 로그 플롯 : semilogy,semilogx(),loglog()
years = df.index.values
counts = df['trans_count'].values

plt.semilogy(years,counts,'o') # semilogy() 함수는 y값을 로그변환하여 출력

poly = np.polyfit(years,np.log(counts),deg=1)
print(poly)   # [ 3.61559210e-01 -7.05783195e+02] 회귀계수
# H(X) =  3.61559210e-01*x - 7.05783195e+02  ,예측방정식(함수)

plt.semilogy(years,np.exp(np.polyval(poly,years)))  # 예측함수의 직선
# plt.semilogy(years,np.exp(3.61559210e-01*years - 7.05783195e+02))
plt.show

# np.polyval(poly,years) = 3.61559210e-01*x - 7.05783195e+02 
# polyfit()으로 얻은 poly내에 기울기와 y절편을 이용하여 예측함수에 x값(years)을
# 넣어 예측값(y,counts,H(x))을 구한다
```

### 분산 플롯(scatter plot) : scatter()함수, 버블차트

```python
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df,gpu,how='outer',left_index=True,right_index=True)
df = df.replace(np.nan,0)
df
```

```python
years = df.index.values # x값
counts = df['trans_count'].values
cnt_log = np.log(counts)   # 로그변환한 y값
gpu_counts = df['gpu_trans_count'].values   # 버블의 크기

# plt.figure(figsize=(12,10))
plt.scatter(years,cnt_log,c= 200*years,
            s = 20 + 200*gpu_counts/gpu_counts.max(),
            alpha = 0.5)  # c : 컬러 , s : 버블의 크기, alpha : 투명도
plt.grid()
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('tarns & gpu counts per years')
plt.show()
```

```python
import matplotlib

matplotlib.reParams['font.family'] = 'Malgun Gothic'   # 맑은 고딕으로 설정
matplotlib.reParams['axes.unicode_minus'] = False     # minus 숫자 표시

plt.title
plt.show()
```

```python
# 범례(legend), 주석(annotation), 예측함수의 직선 그래프를 함께 출력
years = df.index.values # x값
counts = df['trans_count'].values
cnt_log = np.log(counts)   # 로그변환한 y값
gpu_counts = df['gpu_trans_count'].values   # 버블의 크기

# 추가 코드

# 예측함수의 직선 그래프를 함께 출력
poly = np.polyfit(years,cnt_log,deg=1)
plt.plot(years,np.polyval(poly,years),label='poly fit')

# annotation(주석 달기)
gpu_start = gpu.index.values.min() # 1997
y_ann = np.log(df.at[1997,'trans_count'])
ann_str = 'First GPU\n1997'
plt.annotate(ann_str,xy=(gpu_start,y_ann),
             arrowprops=dict(arrowstyle='->'),
             xytext=(-21,+35), textcoords='offset points')

plt.scatter(years,cnt_log,c= 200*years,
            s = 20 + 200*gpu_counts/gpu_counts.max(),
            alpha = 0.5, label = 'Scatter Plot')  # c : 컬러 , s : 버블의 크기, alpha : 투명도

plt.grid()
plt.legend(loc = 'upper left')
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('tarns & gpu counts per years',fontdict={'fontsize':18,'color':'b'})
plt.show()
```

```python
# 지연 플롯(Lag plot) : time series data(시계열 데이터) 분석
# 시계열 데이터에서 x축과 y축의 값을 지연(Lag) 값으로 차이나게 해서
# 시간의 변화에 따른 데이터의 변화 관계를 

from pandas.plotting import lag_plot

# 지연값 : 1년, lag = 1
# x축 : 기준 연도의 trans_count의 로그 값
# y축 : (기준 연도 +1)의 trans_count의 로그 값
lag_plot(np.log(df['trans_count']),lag=1)
plt.grid()
plt.show()
```

```python
# 자기 상관플롯
from pandas.plotting import autocorrelation_plot

print(df.corr())

# x축 : 시간 지연(lag), 1년 단위
# y축 : 상관 계수
autocorrelation_plot(np.log(df['trans_count']))
plt.grid(True)
plt.show()

# 지연값(Lag)이 작으면 상관계수는 크다
# 지연값(Lag)이 크면 상관계수는 작아지고, 0으로 수렴한다
```

### Boxplot(상자 그림)
* pandas
* matplotlib
* seaborn

```python
# pandas
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df,gpu,how='outer',left_index=True,right_index=True)


# 로그변환, 결측치는 0으로
df['trans_count'] = np.log(df['trans_count'].values)
df['gpu_trans_count'] = np.log(df['gpu_trans_count'].values)

df = df.replace(np.nan,0)

df.plot.box()
plt.text(2,df['trans_count'].median(),round(df['gpu_trans_count'].median(),2))
plt.show()
```

```python
# matplotlib
df = df.replace(np.nan,0)
plt.boxplot([df['trans_count'],df['gpu_trans_count']])
plt.show()
```

```python
# seaborn : 결측치가 있어도 출력
import seaborn as sns
sns.boxplot(data=df, order=['trans_count','gpu_trans_count'],
            width=0.15,linewidth=1.2)

plt.show()
```
