### DataFrame의 요소 접근하기 : 읽기와 수정
* (1) df.loc[행번호,열] : 명시적인 index
* (2) df.iloc[행번호,열] : 함묵적인 index, 파이썬 스타일, int형 0 부터 시작
* df.ix[] ; 최신 버전에서 삭제됨

```python
# df.loc[]
df.loc[3,'Country']   # 3번 행, 'Country' 컬럼 ,Andorra
df.loc[:4,'Country']# start : end , end + 1 이 아님을 주의
df.loc[:4,['Country','CountryID','Continent']]
```

```python
# df.iloc[]
df.iloc[3,0]    # 3번 행, 0번 컬럼 ,Andorra
df.iloc[:4,0]   #  start : end + 1
df.iloc[:4,0:3]
df.iloc[:4,[0,2,4]]

df.iloc[:,:-1]   # 마지막 컬럼만 제외, X(독립변수) 추출
df.iloc[:,-1]    # 마지막 컬럼만 추출, Y(종속변수) 추출

# -1 사용 가능
df.iloc[-1,0]
df.iloc[0,-1]

# 스칼라 요소 값의 변경
df.iloc[3,0] = 'Dorra'     # ' Andorra'  ==> 'Dorra'
df.iloc[3,0] = 'Andorra'   # 'Dorra' ==>  'Andorra' 
```
