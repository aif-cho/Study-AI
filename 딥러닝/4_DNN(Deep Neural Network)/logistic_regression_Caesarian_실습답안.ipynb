```python
# logistic_regression_Caesarian.

import tensorflow as tf
import numpy as np
tf.random.set_seed(5)

xy = np.loadtxt('caesarian.csv',delimiter=',',dtype=np.float32)

# train data set
x_data = xy[0:56, 0:-1 ]
y_data = xy[0:56, [-1] ]

x_train = np.array(x_data,dtype=np.float32) # (56, 5)
y_train = np.array(y_data,dtype=np.float32) # (56, 1)
```

```python
# Layer  1
# (56, 5) * (5,5)= (56, 5)
W1 = tf.Variable(tf.random.normal([5,5]), name='weight1')
b1 = tf.Variable(tf.random.normal([5]), name='bias1')

@tf.function
def layer1(X):
    return  tf.nn.relu(tf.matmul(X,W1) + b1 )
```

```python
# layer 2
# (56, 5) * (5,2)= (56, 2)
W2 = tf.Variable(tf.random.normal([5,2]), name='weight2')
b2 = tf.Variable(tf.random.normal([2]), name='bias2')

@tf.function
def layer2(X):
    return  tf.nn.relu(tf.matmul(layer1(X),W2) + b2 )
```

```python
# layer 3
# (56, 2) * (2,1)= (56, 1)
W3 = tf.Variable(tf.random.normal([2,1]), name='weight3')
b3 = tf.Variable(tf.random.normal([1]), name='bias3')
```

```python
# hypothesis 예측 함수(방정식) : 활성화 함수 사용
def hypothesis(X):
    return  tf.sigmoid(tf.matmul(layer2(X),W3) + b3)
```

```python
# 비용함수 : logloss
def cost_func():
    cost = -tf.reduce_mean(y_train*tf.math.log(hypothesis(x_train)) + (1-y_train)*
                       tf.math.log(1-hypothesis(x_train)))
    return cost
```

```python
## 경사 하강법
# learning rate (학습율) 을 0.01로 설정하여  optimizer객체를 생성
# optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.01)
optimizer = tf.keras.optimizers.Adam(lr=0.01)
```

```python
# 학습 시작
print('***** Start Learning!!')
for step in range(10001):
    # cost를 minimize한다
    optimizer.minimize(cost_func, var_list=[W1,b1,W2,W2])
    
    if step % 1000 == 0:
        print('%04d'%step, 'cost: [', cost_func().numpy(), ']  W: ', W1.numpy(), '  b: ', b1.numpy())
        
print('***** Learning Finished!!')
```

```python
# 회귀 계수 출력
print('Weight =', W1.numpy())
print('Bias =', b1.numpy())         
print('Weight =', W2.numpy())
print('Bias =', b3.numpy())   
print('Weight =', W3.numpy())
print('Bias =', b3.numpy())      
```

```python
# accuracy computation (정확도 측정)
# test data set
x_data = xy[56:, :-1 ]
y_data = xy[56:, [-1] ]

x_test = np.array(x_data,dtype=np.float32)
y_test = np.array(y_data,dtype=np.float32)

def predict(X):
    return tf.cast(hypothesis(X) > 0.5, dtype = tf.float32)

accuracy = tf.reduce_mean(tf.cast(tf.equal(predict(x_test),y_test),
                                     dtype = tf.float32))
# print("Hypothesis:\n",hypothesis(x_test).numpy(), 
#       "\nPredict:\n",predict(x_test).numpy())

print("\nAccuracy:",accuracy.numpy()) # Accuracy: 0.45833334

```
